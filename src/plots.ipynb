```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("varying_attr_r_min_0.0_r_max_8.0_N_100_Dattr_2_Dweight_2.txt")
fig, ax = plt.subplots(figsize=(8,6))
for name, group in df.groupby("algo_name"):
    group.plot(x="attr_R",y="mean_F1",yerr="std_F1",label=name,ax=ax)
plt.ylabel("F1-score")
plt.savefig("increase_attr_info.png")
plt.show()
```
